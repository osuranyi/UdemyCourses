## **14. Line chart**

First, the necessary libraries are included:

```python
import numpy as np
import matplotlib.pyplot as plt
```

Next, we generate data. `np.linspace` is used to generate a certain number of evenly spaced values in a numpy array.

```python
x = np.linspace(0,20,1000)
y = np.sin(x) + 0.2 * x
```

Plotting data:

```python
plt.plot(x,y)
```

*Remark:* If not using a notebook, `plt.show()` should be called

Adding labels:

```python
plt.plot(x,y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('my plot'); # semicolon suppresses output
```

## **15. Scatterplot**

First, creating some random data using standard normal distribution. We need an array of (x1,x2) values to have a meaningful scatterplot:

```python
X = np.random.randn(100,2)
plt.scatter(X[:,0],X[:,1]);
```

A more complicated example. Shifting first 50 datapoint by (+3,+3). Also coloring them differently using the `c=` color option.

```python
X = np.random.randn(200,2)
X[:50] += 3
Y = np.zeros(200)
Y[:50] = 1
plt.scatter(X[:,0],X[:,1], c=Y);
```

## **17. Plotting images**

Only considering single images right now. First download an image:

```python
!wget https://dv-website.s3.amazonaws.com/uploads/2018/05/kf_ann_052418.png
```

Open the image in python:

```python
from PIL import Image

im = Image.open('kf_ann_052418.png')
type(im)
```

Notice, that **this is not a numpy array**, but still a PNG! But it's easy to convert it to an array:

```python
arr = np.array(im)
arr
```

All values are integers and between 0 and 255. These are 8-bit unsigned integers ($2^8 = 256$). Check the array shape:

```python
arr.shape
```

Third dimension contains the values of red, green, and blue channels. Let's plot the image:

```python
plt.imshow(arr);
```

This also works with the image object as the argument:

```python
plt.imshow(im);
```

Make it grayscale: take the average of channels:

```python
gray = arr.mean(axis=2)
plt.imshow(gray);
```

This is a heatmap. Let's change it to actual grayscale:

```python
plt.imshow(gray,cmap='gray');
```

## **18. Matplotlib exercise**

* Generate a generalized XOR dataset: `x1 > 0 XOR x2 > 0`
* Plot on a scatterplot with two colors

First generate a set of (x,y) pairs uniformly distributed in the $|x|<1$, $|y|<1$

```python
N = 1000
X = (2*np.random.random((N,2)) - 1)
plt.scatter(X[:,0],X[:,1]);
```

Calculate true and false values:

```python
T = np.zeros(N)
T = (X[:,0] > 0) ^ (X[:,1] > 0)
plt.scatter(X[:,0],X[:,1],c=T);
```
